```python
import numpy as np
import matplotlib.pyplot as plt
import random 
```

```python
x = np.array([0,6])
y = np.array([10,250])

v = [1,2,3,5,6,7]
z = [10,20,30,100,50,200]

```

```python
print(y)
```

```python
plt.plot(v,z)
plt.show
```

```python
plt.plot(x,y,color="y",marker="o",linestyle="-")
plt.title("Title")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
#Multiple line charts

x1 = [1,2,3,4,5]
y1 = [10,50,30,40,80]
y2 = [20,30,40,50,60]

plt.plot(x1,y1,label="Trend")
plt.plot(x1,y2,label="Expected")

plt.xlabel("Months")
plt.ylabel("Values")
plt.title("Title")

plt.legend()
plt.show()
```

```python
#Bar chart

x = np.array(["A","B"])
y = np.array([10,100])

plt.bar(x,y)
```
